```python
from sentence_transformers import SentenceTransformer
import torch
import torch.nn.functional as F

encoder = SentenceTransformer("BAAI/bge-large-en-v1.5")

# Pre-encode KG relations once
kg_relations = ['year_of_birth', 'place_of_birth', 'date_of_birth', 'birthdate', 'birth_year', 'country_of_birth',
                'located_in_area', 'place_of_death', 'date_of_death', 'death_place', 'death_year', 'country_of_death',
                'educated_at', 'alma_mater', 'member_of', 'occupation', 'position_held', 'religion', 'ethnicity',
                'citizenship', 'language_spoken', 'award_received', 'notable_work']
kg_embeds = encoder.encode(kg_relations, normalize_embeddings=True)

def batch_Sim(r_query: str, top_k=5):
    """
    Compute top-k most similar relations in KG to query relation.
    """
    query_embed = encoder.encode([r_query], normalize_embeddings=True)
    scores = query_embed @ kg_embeds.T
    top_indices = torch.topk(torch.tensor(scores), top_k).indices.tolist()[0]
    return [(kg_relations[i], scores[0][i].item()) for i in top_indices]

print(batch_Sim("birth", top_k=3))
```
